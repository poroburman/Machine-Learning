```python
# This script uses ANN Regression model to predict the energy output of Power Plant.
```

```python
# 1. DATA PREPROCESSING
```

```python
# Import libraries
import os
import numpy as np
import pandas as pd
import tensorflow as tf

# set the working directory
os.chdir('/Users/poromendro/Desktop')

# import the dataset
dataset = pd.read_excel('Folds5x2_pp.xlsx')

# set the independent and dependent variables
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

# split dataset into training and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# 2. BUILD THE ANN
```

```python
# create ann instance
ann = tf.keras.models.Sequential()

# add input layer and first hidden layer
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))

# add second hidden layer
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))

# add the output layer
ann.add(tf.keras.layers.Dense(units = 1))
```

```python
# 3. TRAINING THE ANN
```

```python
# compiling the ANN
ann.compile(optimizer = 'adam', loss = 'mean_squared_error')

# training the ANN on the training set
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

```python
# 4. MAKE A PREDICTION
```

```python
# Predict test set results
y_pred = ann.predict(X_test)
np.set_printoptions(precision=2)
print(f"1. np array with predicted values in the first column and test values in the second column \n\n {np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)}")
```
